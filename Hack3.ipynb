```python
import numpy as np 
import pandas as pd 
import os
```

```python
raw_data = pd.read_excel('Master_Data.xlsx')
```

```python
raw_data.info()
```

```python
raw_data = raw_data.dropna(how='any', axis = 0)
```

```python
raw_data['Gender'] = raw_data['Gender'].map({'Male':0, 'Female':1})
raw_data['Dataset'] = raw_data['Dataset'].map({2:0, 1:1}) 

raw_data
```

```python
data_corr = raw_data.drop(['Gender', 'Dataset'], axis = 1)
data_corr.corr()
```

```python
samples_count = raw_data.shape[0]

data_train_count = int(0.8*samples_count) 
data_test_count = samples_count - data_train_count

data_train_data = raw_data[:data_train_count]
data_test_data = raw_data[data_train_count:]
```

```python
print(data_train_count)
print(data_test_count)
```

```python
import statsmodels.api as sm
```

```python
X1_train = data_train_data[['Age','Gender','Total_Bilirubin','Direct_Bilirubin','Alkaline_Phosphotase','Alamine_Aminotransferase',
              'Aspartate_Aminotransferase','Total_Protiens','Albumin', 'Albumin_and_Globulin_Ratio' ]]



X_train= sm.add_constant(X1_train)

y_train = data_train_data['Dataset']


X1_test = data_test_data[['Age','Gender','Total_Bilirubin','Direct_Bilirubin','Alkaline_Phosphotase','Alamine_Aminotransferase',
              'Aspartate_Aminotransferase','Total_Protiens','Albumin', 'Albumin_and_Globulin_Ratio' ]]

X_test = sm.add_constant(X1_test)

y_test = data_test_data['Dataset']
```

```python
reg_log = sm.Logit(y_train,X_train)
result_log = reg_log.fit()

result_log.summary2()


```

```python

result_log.predict()
result_log.pred_table()
```

```python

X_test.head()
```

```python
X_train.head()
```

```python
logit_predicted = result_log.predict(X_test)
```

```python
bins = np.array([0,0.5,1])
cm_log = np.histogram2d(y_test, logit_predicted, bins = bins)[0]
logit_accuracy = (cm_log[0,0] + cm_log[1,1])/cm_log.sum()
```

```python
print('Confusion Matrix (Logit): \n', cm_log)
print('---------------------------------------------')
print('Accuracy (Logit): \n', round(logit_accuracy*100,2), '%')
```

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
from sklearn.linear_model import LogisticRegression
```

```python
### Gaussian Naive Bayes

from sklearn.naive_bayes import GaussianNB

gaussian = GaussianNB()
result_gauss = gaussian.fit(X_train,y_train)

gauss_predicted = gaussian.predict(X_test)
```

```python
print('Confusion Matrix (Gaussian): \n', confusion_matrix(y_test,gauss_predicted))
print('---------------------------------------------')
print('Accuracy (Gaussian): \n', round(accuracy_score(y_test, gauss_predicted)*100,2), '%')
```

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
from sklearn.linear_model import LogisticRegression

logreg = LogisticRegression()
logreg.fit(X_train,y_train)

log_predicted = logreg.predict(X_test)
```

```python
print('Confusion Matrix (Logistic Reg): \n', confusion_matrix(y_test,log_predicted))
print('---------------------------------------------')
print('Accuracy (Logistict Reg): \n', round(accuracy_score(y_test, log_predicted)*100,2), '%')
```
